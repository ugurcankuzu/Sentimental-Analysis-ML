```python
import numpy as np
import pandas as pd
import re

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

import nltk
from nltk import word_tokenize

!pip install pickle5
import pickle5 as pk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
trainData = pd.read_csv("/content/twitter_training.csv",header=None)
trainData
```

```python
trainData.columns = ["id","takenFrom","type","text"]
trainData
```

```python
trainDup = trainData
```

```python
#Text Format Section

trainDup["lower"] = trainDup.text.str.lower() #Text lowercase convert
trainDup["lower"] = [str(data) for data in trainDup.lower] #Text string convert
trainDup["lower"] = trainDup.lower.apply(lambda x: re.sub('[^A-Za-z0-9 ]+', ' ', x)) #Text filtering using RegEx

trainDup
```

```python
#Text Split Section
textToken = [word_tokenize(str(word)) for word in trainDup.lower]

tokenCount = [item for text in textToken for item in text]

print("Token Count is : {}".format(len(set(tokenCount))))
```

```python
#Ineffective words in english for model

stopWords = nltk.corpus.stopwords.words('english')
stopWords[:10] #Example
```

```python
#Vectorizing Texts

vectorizer = CountVectorizer(
    tokenizer = word_tokenize,
    stop_words = stopWords,
    ngram_range = (1,1)
)
```

```python
#Split Dataset

Xtrain,Xtest = train_test_split(trainDup,test_size=0.2,shuffle=True)
```

```python
XtrainVector = vectorizer.fit_transform(Xtrain.lower)
XtestVector = vectorizer.transform(Xtest.lower)
```

```python
YtrainVector = Xtrain.type
YtestVector = Xtest.type
```

```python
XtestVector.shape
```

```python
Xtest
```

```python
latestModel = None
for i in np.arange(0.1,1.0,0.1):
  accuracy = 0
  model = LogisticRegression(C=i,solver="liblinear",max_iter=200)
  model.fit(XtrainVector,YtrainVector)
  prediction = model.predict(XtestVector)
  calc = accuracy_score(YtestVector,prediction)*100

  print("For C value: {}Accuracy Value: {}".format(i,calc))
  if(calc > accuracy):
    accuracy = calc
    latestModel = model
```

```python
with open("model.pickle","wb") as f:
  pk.dump(latestModel,f)
  
```
